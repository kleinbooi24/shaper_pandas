```python
#PULL THE DATA TO BE USED

#We want to count the number of transections paid via MPESA send money

#ALL TRANSECTIONS
#df = money_df["IS_PURCHASE_PAID_VIA_MPESA_SEND_MONEY"].count()
#print(f"Total including True/ False is {df}")
#ONLY The transections with TRUE \
import pandas as pd 
#import matplotlib as plt
import matplotlib.pyplot as plt
money_df = pd.read_csv("Money.csv")

paid_via_mpesa_count_True = money_df[money_df["IS_PURCHASE_PAID_VIA_MPESA_SEND_MONEY"] == True]
paid_via_mpesa_count_False = money_df[money_df["IS_PURCHASE_PAID_VIA_MPESA_SEND_MONEY"] == False]
#length = len(paid_via_mpesa_count_True)
#length

#compare= [paid_via_mpesa_count_True["IS_PURCHASE_PAID_VIA_MPESA_SEND_MONEY"], paid_via_mpesa_count_False["IS_PURCHASE_PAID_VIA_MPESA_SEND_MONEY"]]
#compare

paid_counts = money_df["IS_PURCHASE_PAID_VIA_MPESA_SEND_MONEY"].value_counts() # value_counts() - counts the number of unique values

print(paid_counts)

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(paid_counts, labels=paid_counts.index, autopct='%1.1f%%', startangle=140) # I can change autopct to autopct='%1.1f%%' to show in percentahes
plt.title('MPESA Transaction Payment Method Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()



```
